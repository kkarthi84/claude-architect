```python
from dotenv import load_dotenv
import os
import anthropic

load_dotenv()

# Create client

client = anthropic.Anthropic(api_key=os.environ.get("ANTRHOPIC_API_KEY"))
model = "claude-haiku-4-5"


# Make an initial list of message

messages = []

# Add user message
def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

# Add asst message
def add_asst_message(messages, text):
    asst_msg = {"role": "assistant", "content": text}
    messages.append(asst_msg)
```

```python
# Chat
def chat(messages):
    stream = client.messages.create(
        model=model,
        max_tokens=500,
        messages=messages,
        stream=True
    )

    for event in stream:
        print(event)
```

```python
# Stream chat
def streamChat(messages):
    with client.messages.stream(
        model=model,
        max_tokens=500,
        messages=messages
    ) as stream:
        for text in stream.text_stream:
            print(text, end="")
            #pass

    # stream.get_final_message() # To store in object or save in db
```
