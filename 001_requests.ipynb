```python
#Load Env Variables

from dotenv import load_dotenv
import os
import anthropic

load_dotenv()
```

```python
#Create an API Client


client = anthropic.Anthropic(api_key=os.environ.get("ANTRHOPIC_API_KEY"))
model = "claude-sonnet-4-6"
```

```python
# Conversational Messages

def add_user_message(messages, text):
    user_message = {"role":"user", "content": text}
    messages.append(user_message)

def add_asst_message(messages, text):
    asst_message = {"role":"assistant", "content": text}
    messages.append(asst_message)

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages
    )
   
    return message.content[0].text

```

```python
# Make a starting list of messages

messages = []

# Add in the initial message

add_user_message(messages, "what is backbone for trading/investing?")

# Take answers and add it to as an asst message
answer = chat(messages)
add_asst_message(messages, answer)

# Add followup question
add_user_message(messages, "Tell me more about Risk Managment like guide line for success")

# Call chat again with the list of messages to get final answer
answer = chat(messages)

print(answer)

```
